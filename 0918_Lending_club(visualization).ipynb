* loan_amnt             : 차용인이 신청한 대출의 나열된 금액
* term                  : 대출에 대한 지불 횟수(월, 36 or 60)
* int_rate              : 대출 이자율
* installment           : 월별 지불금
* grade                 : 대출 등급
* sub_grade             : 대출 등급의 sub 등급
* emp_title             : 대출자 직함
* emp_length            : 고용 기간(년) (1 > , 1 < 10 , < 10)
* home_ownership        : 주택 소유 상태(임대, 소유, 모기지, 기타)
* annual_inc            : 연간 소득
* verification_status   : 소득 확인 여부
* issue_d               : 대출 자금이 조달된 월
* loan_status           : 대출 현황!!
* purpose               : 대출 요청을 위해 차용인이 제공하는 범주 (대출 목적)(debt_consolidation 부채 통합)

* title                 : 차용인이 제공한 대출 제목 - 쓸모 없음
* dti                   : LC 대출을 제외한 부채에 대한 월별 부채 총액 / 차용인의 월 소득
* earliest_cr_line      : 신용 한도가 개설된 날
* open_acc              : 열린 한도 수
* pub_rec               : 경멸적인 공개 기록의 수
* revol_bal             : 총 신용 회전 잔액
* revol_util            : 사용 가능 회전 잔액에 비해 차용인이 사용하고 있는 크레딧 금액
* total_acc             : 총 신용 한도 수
* initial_list_status   : 대출의 초기 목록 상태 (W, F)
* application_type      : 개인 신청인지 공동 신청인지
* mort_acc              : 모기지 계좌 수
* pub_rec_bankruptcies  : 공공 기록 파산 건수
* address               : 주소

```python
import numpy as np
import random

def rand_color():
    return '#' + "".join([random.choice('0123456789ABCDEF') for _ in range(6)])

numerical_columns = df.select_dtypes(include=[np.number]).columns.tolist()
num_cols = len(numerical_columns)
num_rows = num_cols // 4
num_rows += num_rows % 4
position = range(1, num_cols+1)



fig = plt.figure(figsize=(16, num_rows * 4))

#서브플롯으로 boxplot 그리기
fig = plt.figure(figsize=(16, num_rows*4))
for k, col in zip(position, numerical_columns):
  ax = fig.add_subplot(num_rows, 5, k)
  # boxplot
  sns.boxplot(df[col], color=rand_color(), ax=ax)
  ax.set_title(col)

plt.tight_layout()
plt.show()

'''
dti, pub_rec, revol_util
특히 높은 이상치 삭제해야 할까
'''
```
